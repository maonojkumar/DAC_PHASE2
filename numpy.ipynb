import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```python
data = pd.read_csv('/content/DDW_B06SC_3300_State_TAMIL_NADU-2011.csv')
```

```python
print(data.head())

```

```python
print(data.info())

```

```python
print(data.describe())
```

```python
data = data.dropna()
data['Age group'] = data['Age group'].astype('category')
```

```python
plt.hist(data['Age group'], bins=10)
plt.xlabel('Age group')
plt.ylabel('Frequency')
plt.title('Age Group Distribution')
plt.show()
```

```python
data.to_csv('preprocessed_data.csv', index=False)

```

```python
age_group_counts = data['Age group'].value_counts()
age_group_counts.plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of Individuals in Each Age Group')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(data['Worked for 3 months or more but less than 6 months -  Persons'], bins=20)
plt.xlabel('Number of Persons')
plt.ylabel('Frequency')
plt.title('Distribution of Worked for 3-6 Months - Persons')
plt.show()
```

```python
plt.hist(data['Worked for 3 months or more but less than 6 months - Males'], bins=20)
plt.xlabel('Number of Persons')
plt.ylabel('Frequency')
plt.title('Distribution of Worked for 3-6 Months - Persons')
plt.show()
```

```python
plt.hist(data['Worked for 3 months or more but less than 6 months - Females'], bins=20)
plt.xlabel('Number of Persons')
plt.ylabel('Frequency')
plt.title('Distribution of Worked for 3-6 Months - Persons')
plt.show()
```

```python
plt.boxplot(data['Worked for 3 months or more but less than 6 months - Females'])
plt.title('Box Plot of Work Duration (3-6 months)')
plt.ylabel('Number of Persons')
plt.show()
```

```python
correlation_matrix = data.corr()
# Use Seaborn for a heatmap
import seaborn as sns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Group the data by 'Age group' and calculate the mean and median work duration
age_group_stats = data.groupby('Age group')['Worked for 3 months or more but less than 6 months - Females'].agg(['mean', 'median'])

# Reset the index for a cleaner representation
age_group_stats = age_group_stats.reset_index()

# Print the results
print(age_group_stats)
```
